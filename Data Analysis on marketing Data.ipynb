IQR

```python
q1=data['Income'].quantile(0.25)
q3=data['Income'].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR=q3-q1
```

```python
IQR
```

```python
lower_bound=q1 - 1.5*IQR
upper_bound=q3 + 1.5*IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
cleaned_data=data[(data['Income']<upper_bound) & (data['Income']>lower_bound)]
```

```python
cleaned_data['Income'].plot(kind='kde')
```

```python
cleaned_data['Income'].plot(kind='hist')
```

```python
cleaned_data['Income'].plot(kind='box')
```
